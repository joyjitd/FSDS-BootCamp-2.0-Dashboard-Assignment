1.	Write a Python Program to Display Fibonacci Sequence Using Recursion?

```python
# Python program to display the Fibonacci sequence

def recur_fibo(n): # Recursion function calls itself 
   if n <= 1:
       return n
   else:
       return(recur_fibo(n-1) + recur_fibo(n-2))

nterms = 10

# check if the number of terms is valid
if nterms <= 0:
   print("Plese enter a positive integer")
else:
   print("Fibonacci sequence:")
   for i in range(nterms):
       print(recur_fibo(i))
```

2.	Write a Python Program to Find Factorial of Number Using Recursion?

```python
def factorial(x):
    """This is a recursive function
    to find the factorial of an integer"""

    if x == 1:
        return 1
    else:
        return (x * factorial(x-1))


num = int(input("Enter the number to find the factorial of that number using recursion: "))
#  print("The factorial of", num, "is", factorial(num))
#num = 3
print("The factorial of", num, "is", factorial(num))
```

3.	Write a Python Program to calculate your Body Mass Index?

```python
the_height = float(input("Enter the height in cm: "))  
the_weight = float(input("Enter the weight in kg: "))  
the_BMI = the_weight / (the_height/100)**2  # Using the formula for Body mass Index. 
print("Your Body Mass Index is", the_BMI)  
```

4.	Write a Python Program to calculate the natural logarithm of any number?

```python
# Python code to demonstrate the working of
# log(a,Base) 

import math # Using the math library

# Printing the log base e of 14
print ("Natural logarithm of 14 is : ", end="")
print (math.log(14))

# Printing the log base 5 of 14
print ("Logarithm base 5 of 14 is : ", end="")
print (math.log(14,5))
```

5.	Write a Python Program for cube sum of first n natural numbers?

```python
# Program  print cube sum of first n natural numbers
n=int(input("Enter the natural number:- "))
s=0
# iterating loop up to given number n
print(f"The following are the cube of the first {n} natural numbers: ")
for i in range(1,n+1):
	# adding cube sum using pow() function...pow function heads to find value to the power of 
  s=s+pow(i,3) # Using inbuild function of pow()
  print(s)
```
