3.	Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

```python
dec = 145
print("The decimal value of", dec, "is:")
print(bin(dec), "in binary.")
print(oct(dec), "in octal.")
print(hex(dec), "in hexadecimal.")
```

5.	Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

```python
def add(x,y):
  return x+y
def sub(x,y):
  return x-y
def mul(x,y):
  return x*y
def div(x,y):
  return x/y
print("""Select the operation you would like to do:
1.Addition
2.Subtraction
3.Multiplication
4.Division
""")
while True: # True is given so that loops continues to run untill user stops it. 
  options=input("Enter your operation choice (1,2,3 or 4): ")
  if options in ('1', '2', '3', '4'):
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    if options == '1':
            print(num1, "+", num2, "=", add(num1, num2))
    elif options == '2':
            print(num1, "-", num2, "=", sub(num1, num2))
    elif options == '3':
            print(num1, "*", num2, "=", mul(num1, num2))
    elif options == '4':
            print(num1, "/", num2, "=", div(num1, num2))
# break the while loop if answer is no
    next_calculation = input("Do you want to do another calculation? (yes/no): ")
    if next_calculation == "no":
      break
  else:
    print("Invalid input. Please choose the options: Enter your operation choice (1,2,3 or 4) ")
```
