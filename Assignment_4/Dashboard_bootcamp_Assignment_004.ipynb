# 10. What is difference between the list methods append() and insert()?
#Answer
In Python, the difference between the append() and extend() method is that:
- The append() method adds a single element to the end of a list.
- The extend() method adds multiple items.

```python
a = [1, 2, 3]
a.append(['JOJO',123, 'Python']) #append() method adds a single element to the end of a list.
print(a) #OUTPUT [[1, 2, 3, ['JOJO', 123, 'Python']]
```

```python
a = [1, 2, 3]
a.extend(['JOJO',123, 'Python']) # The extend() method adds multiple items
print(a) #OUTPUT [1, 2, 3, 'JOJO', 123, 'Python']  
```

# 17. How do you distinguish between copy.copy() and copy.deepcopy()?
#Answer
The copy.copy() function will do a shallow copy of a list,
The copy.deepcopy() function will do a deep copy of a list. only copy.deepcopy() will duplicate any lists inside the list

```python
lst1 = [1,2,3]
lst2 = lst1
mry= [id(lst1),id(lst2)] # lst2 = lst1 = operation memory location is same 
mry
```

```python
lst1 = [1,2,3]
lst2 = lst1.copy()  
mry= [id(lst1),id(lst2)] # # lst2 = lst1.copy() = operation memory location is differnce when using shallow copy() 
mry
```

```python
# DeepCopy mained is helpfull in nested list
import copy # copy is a library 
lt1 = [1,2,3,4]
lt2 =copy.deepcopy(lt1)
lt2[1]=100
dis= [lt1,lt2]
dis 
#Normal list copy  == deepcopy in the case of non nested list. 
```

```python
# DeepCopy mained is helpfull in nested list
import copy 
lt1 = [[1,2,3],[9,8,7]]
lt2 =copy.deepcopy(lt1)
mry= [id(lt1),id(lt2)]
mry # #Normal list copy  not equal to deepcopy in the case of nested list. 
```
